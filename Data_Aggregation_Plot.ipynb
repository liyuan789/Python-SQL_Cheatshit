(b) (20 points) Use the Pandas "groupby" method to obtain the mean and the standard deviation of "Results" for each drug, separately for males and females. Save the results in a new data frame. Your new data frame should have exactly the columns "Gender", "Treatment", "Mean Result", "Standard Deviation" (in this order). Display your new data frame to the console so that the grader can see it.

```python
mean = data.groupby(['Treatment','Gender']).mean().Result
std = data.groupby(['Treatment','Gender']).std().Result          # Index Columns included
df = pd.DataFrame({'Mean Result':mean, 'Standard Deviation':std})# create a new dataframe
df.reset_index(inplace = True)                                   # delete index 
df
```

[Practice 2] Read the data in the file "Scores.csv" (which you can download from Canvas) into Python as a Pandas Data Frame. Suppose that the data in this file are the scores of different students (identifyable through their "ID") on different assignments. The assignments include six Homework assignments, eleven Labs, ten Quizzes, two midterms, and a final project. 

(a) Write Python code to automatically extract all columns that start with ("HW", dont' extract them by position or by explicit name). Drop the lowest HW score and compute the percentage of points for each student after dropping the lowest score. Treat missing scores as zeros.  

Write function that takes student ID as its input and returns the total HW score (as a number between 0 and 100) for that student. 

Example: HW_score(49865) = 99.4, HW_score(60921) = 54.0

```python
file = "Scores.csv"
data = pd.read_csv(file)
data.head()

def HW_score(ID):
    ID = float(ID)
    data.set_index(['ID'], inplace = True)
    data.fillna(0)

    # create sub_dataframe
    HW = []
    for column in data.columns:
        if column[:2] == 'HW':
            HW.append(column)     
    HW_data = data[HW]
    
    # data manipulation
    HW_min = HW_data.min(axis=1)   # 这些都是局部变量
    HW_sum = HW_data.sum(axis=1)   # 出了 function 就不存在了
    HW_score = (HW_sum - HW_min)/5 # def function 是为了一种目的
    return HW_score.loc[ID]

HW_score(49865)
```

(b) Find the student ID(s) of the student(s) with HW scores in the F range (below 60).  

```python
data = pd.read_csv(file)
data.set_index(['ID'], inplace = True)       # indexing first
data = data.fillna(0)                        # fill missing values with zeros

HW = []
for column in data.columns:
    if column[:2] == 'HW':
        HW.append(column)        
HW_data = data[HW]

HW_min = HW_data.min(axis=1)
HW_sum = HW_data.sum(axis=1)
HW_score = (HW_sum - HW_min)/5 

idx = HW_score<60                            # True or False
data[idx].index.values[0]                    
```
